```python
# Players who have registered and bought prata in first 10 days
df_first_10_days = df[df['day_after_registration']<=10]
df_first_10_days
```

```python
df_after_10_days = df[df['day_after_registration']>10]
df_after_10_days
```

```python
targets_df = df_after_10_days.groupby('player_identifier').sum().assign(prata_bought = 1)
targets_df
```

```python
# Feature engineeering
# Make recency from the data
max_date = df_first_10_days['order_date'].max()
recency_features_df = df_first_10_days[['player_identifier', 'order_date']] \
                        .groupby('player_identifier') \
                        .apply(lambda x: (x['order_date'].max() - max_date) / pd.to_timedelta(1,'day')) \
                        .to_frame()\
                        .set_axis(['recency'], axis = 1)
recency_features_df.info() 
```

```python
frecency_features_df = df_first_10_days[['player_identifier', 'order_date']] \
                        .groupby('player_identifier').count().set_axis(['frequency'], axis=1)
frecency_features_df 
```

```python
price_features_df = df_first_10_days.groupby('player_identifier')['prata_spent'].sum().to_frame().set_axis(['prata_sum'], axis=1)
price_features_df
```

```python
# Combine features
features_df = pd.concat([recency_features_df, frecency_features_df, price_features_df], axis = 1) \
                .merge(targets_df,\
                       left_index  = True, \
                       right_index = True, \
                       how = "left")
features_df = features_df.replace(np.nan,0)
# features_df = features_df.drop(['day_after_registration', 'prata_spent', 'prata_bought'], axis = 1)
features_df.info()
```

```python
features_df
```

```python
X = features_df[['recency', 'frequency', 'prata_sum']]
y = features_df['prata_spent']
```

```python
xgb_reg = XGBRegressor(objective = "reg:squarederror", random_state = 1)
xgb_reg_model = GridSearchCV(estimator = xgb_reg, param_grid = dict(learning_rate = [0.01,0.1,0.3,0.5]), \
                             scoring = 'neg_mean_absolute_error',
                             refit = True,
                             cv = 5)
```

```python
xgb_reg_model.fit(X,y)
```

```python
xgb_reg_model.best_score_
```

```python
xgb_reg_model.best_params_
```

```python
xgb_reg_model.best_estimator_
```

```python
predictions_reg = xgb_reg_model.predict(X)
predictions_reg
```

# Third Approch

Calculating players that registered and spent in less than 10 days


```python
df_10_days = df[(df.day_after_registration < 10)].reset_index(drop =True)
df_10_days
```

```python
# Finding the unique player ids
cust_ids = df_10_days.player_identifier.unique() 
len(cust_ids)
```

```python
# Find the rest of the data of these customers
df_data_10_days = df[df.player_identifier.isin(cust_ids)]
df_data_10_days
```

```python
# Understanding the min and max date of the data
print(df_data_10_days.order_date.min())
print(df_data_10_days.order_date.max())

print(df_data_10_days.date_registered.min())
print(df_data_10_days.date_registered.max())
```

```python
# filtering the orders for less than 2 years
tx_less_2yrs = df_data_10_days[(df_data_10_days.order_date <= '2019-11-30')]
tx_less_2yrs
```

```python
# filtering the data for more than 2 years
tx_greater_2yrs = df_data_10_days[(df_data_10_days.order_date >= '2019-12-1')]
tx_greater_2yrs
```

### Feature Engineering

Finding Mninmum date of transaction greater than 2 years

```python
tx_greater_2yrs_first_purchase = tx_greater_2yrs.groupby('player_identifier').order_date.min().reset_index()
```

```python
tx_greater_2yrs_first_purchase.columns = ['player_identifier', 'MinPurchaseDate']
```

```python
tx_greater_2yrs_first_purchase.head()
```

```python
tx_less_2yrs_last_purchase = tx_less_2yrs.groupby('player_identifier').order_date.max().reset_index()
tx_less_2yrs_last_purchase.columns = ['player_identifier', 'MaxPurchaseDate']
tx_less_2yrs_last_purchase.head()
```

### Recency 

```python
tx_less_2yrs_last_purchase['Recency'] = (tx_less_2yrs_last_purchase['MaxPurchaseDate'].max() \
                                        - (tx_less_2yrs_last_purchase['MaxPurchaseDate'])) /pd.to_timedelta(1,'day')
```

```python
tx_less_2yrs_last_purchase
```

```python
sse={}
tx_recency = tx_less_2yrs_last_purchase[['Recency']]
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, max_iter=1000).fit(tx_recency)
    tx_recency["clusters"] = kmeans.labels_
    sse[k] = kmeans.inertia_ 
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of cluster")
plt.show()
```
